```python

#df['Review Rating'].fillna(df['Review Rating'].median(), inplace=True)
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x : x.fillna(x.median()) )
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={"purchase_amount_(usd)": "purchase_amount"})
```

```python
df.columns
```

```python
labels =["Young Adult","Adult","Middle Age","Senior"]
df["age_group"] =  pd.qcut(df["age"],q=4,labels =labels)
```

```python
df["age_group"].value_counts() 
```

```python
df["age_group"].value_counts()        # should be ~975 each
df.groupby("age_group")["age"].agg(["min","max","count"])  # see actual ranges
```

```python
df["frequency_of_purchases"].unique()
```

```python
#create a purchase_frequecy_days:
purchase_mapping  = {'Fortnightly' : 14,'Weekly' :7,'Annually':365,'Quarterly' :90,
      'Bi-Weekly' : 14, 'Monthly' : 30, 'Every 3 Months' : 90}

df["purchase_frequecy_days"] = df["frequency_of_purchases"].map(purchase_mapping)
```
